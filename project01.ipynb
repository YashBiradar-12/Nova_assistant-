```python
from pathlib import Path

memory_file = Path('assistant_memory.txt')


def load_memory(path):
    if path.exists():
        return [line.strip() for line in path.read_text(encoding='utf-8').splitlines() if line.strip()]
    return []


def save_memory(path, memories):
    recent_memories = memories[-10:]
    path.write_text('\n'.join(recent_memories), encoding='utf-8')


memories = load_memory(memory_file)
if memories:
    memory_context = '\n'.join(f'- {m}' for m in memories)
else:
    memory_context = 'No previous memories.'

user_prompt = input('Enter your prompt: ')
full_prompt = (
    f"Conversation memory:\n{memory_context}\n\n"
    f"User: {user_prompt}\n"
    "Please respond humorously and keep your answer very concise, suitable for speaking aloud in under 30 seconds."
)
response = model.generate_content(full_prompt)
print('Model Response:')
print(response.text)

memories.append(f'User: {user_prompt}')
memories.append(f'Assistant: {response.text}')
save_memory(memory_file, memories)

text_from_model = response.text
tts = gTTS(text_from_model, lang='en')
tts.save('model_response.mp3')
Audio('model_response.mp3', autoplay=True)
```
